```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
from sklearn.model_selection import KFold, cross_val_score
```

```python
region_one = pd.read_csv('Region_One', delimiter=',')
region_one = region_one.drop(['ID'], axis=1)
region_one
```

```python
region_two = pd.read_csv('Region_Two', delimiter=',')
region_two = region_two.drop(['ID'], axis=1)
region_two
```

```python
region_three = pd.read_csv('Region_Three', delimiter=',')
region_three = region_three.drop(['ID'], axis=1)
region_three
```

```python
region_four = pd.read_csv('Region_Four', delimiter=',')
region_four
```

```python
region_five = pd.read_csv('Region_Five', delimiter=',')
region_five
```

```python
region_six = pd.read_csv('Region_Six', delimiter=',')
region_six = region_six.drop(['ID'], axis=1)
region_six
```

```python
region_seven = pd.read_csv('Region_Seven', delimiter=',')
region_seven = region_seven.drop(['ID'], axis=1)
region_seven
```

```python
data_set = pd.concat([region_one, region_two, region_three, region_seven, region_five, region_six, region_four], ignore_index=True)
print(data_set)
```

```python
features = data_set.drop(['label'], axis=1)
target = data_set['label']
print(features)
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3,random_state=109)
```

```python
SVM = svm.SVC(kernel='rbf', gamma=0.001)
SVM.fit(X_train, y_train)
y_pred = SVM.predict(X_test)
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
```

```python
KNN = KNeighborsClassifier(n_neighbors=13)
KNN.fit(X_train, y_train)
y_pred = KNN.predict(X_test)
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
```

```python
mlp = MLPClassifier(activation='relu', hidden_layer_sizes=(5,5,5), max_iter=3000, random_state=0)
mlp.fit(X_train, y_train)
y_pred = mlp.predict(X_test)
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
```

```python
rf = RandomForestClassifier(n_estimators=200)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
```
